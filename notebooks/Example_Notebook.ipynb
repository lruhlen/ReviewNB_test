```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
plt.style.use('ggplot')
```

```python
x = np.linspace(0, 10, 1000)
```

```python
y = np.sin(x**2) * np.exp(-1.0 * x)
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('X vs Y')
```

```python
plt.plot(x, y**2, c='b')
plt.xlabel('X')
plt.ylabel(r'$Y^2$')
plt.title(r'$X  vs. Y^2$')
```

```python
x = np.arange(0,100)
y = np.random.randint(1,10000, x.size)
plt.plot(x,y, linestyle='none', marker='.',c='g')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Noise')
```
